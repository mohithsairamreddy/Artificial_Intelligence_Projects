# **House Price Predection**


**Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
pre_process = preprocessing.StandardScaler()
```

**Importing Data**

```python
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine_Learning/USA_Housing.csv')
data.head(2)
```

**Analyzing Data**

Finding for Null values

```python
data.isna().sum()
```

Removing Address

```python
data.drop('Address',axis=1,inplace=True)
```

```python
data.head(5)
```

Data to Array

```python
x=data[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y=data['Price']
```

Data Trasformation

**Data Splitting**

Splitting for Test and Train

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state= 2)
```

**Linear Regression**

Importing Libraries

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()


```

Linear Regression Appplication

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
preds = lr.predict(x_test)
```

Training set Results

```python
x_test.size
```

```python
y_test.size
```

Mean Square Error

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, lr.predict(x_test))
mse=np.sqrt(mse)
print('Mean_Squared_Error :' ,mse)

```

R value

```python
r_value=lr.score(x_test,y_test)
print('r_square_value :',r_value)
```
