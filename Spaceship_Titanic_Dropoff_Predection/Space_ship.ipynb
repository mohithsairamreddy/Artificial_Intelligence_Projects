Import Libraries and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pandas import read_csv
train_path=('/content/drive/MyDrive/Colab Notebooks/competition/train.csv')
test_path=('/content/drive/MyDrive/Colab Notebooks/competition/test.csv')
train=read_csv(train_path)
test=read_csv(test_path)
print(train.head(4))
```

Drop Values

```python
train.drop(['PassengerId','Name'],axis=1,inplace=True)
```

Null Values

```python
train.isna().sum()
```

Info

```python
train.info()
```

DataFrame

```python
df=pd.DataFrame(train)
```

Drop Null Values

```python
df=df.dropna()
```

```python
df.isna().sum()
```

Cabin Column Splitting

```python
df[['Deck','Num','Side']]=df['Cabin'].str.split('/',expand=True)
df.drop(['Cabin'],axis=1,inplace=True)
```

Int to text

```python
df['CryoSleep']=df["CryoSleep"].astype(int)
df['VIP']=df["VIP"].astype(int)
df['Side']=df['Side'].map({'P':1,'S':2}).astype('int8')
df['Deck']=df['Deck'].map({'B':1, 'F':2, 'A':3, 'G':4, 'E':5, 'D':6, 'C':7, 'T':8}).astype('int8')
df['Destination']=df['Destination'].map({'TRAPPIST-1e':1, 'PSO J318.5-22':2, '55 Cancri e':3}).astype('int8')
df['HomePlanet']=df['HomePlanet'].map({'Europa':1,'Earth':2,'Mars':3}).astype('int8')
```

```python
df['Transported']=df['Transported'].unique
```

```python
print(df)
```

```python
x=df[['HomePlanet','CryoSleep','Destination','Age','VIP','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck','Deck','Num','Side']]
y=df['Transported']
```

```python
df.info()
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
pedict=dtc.fit(x, y)
```

**R Value**

```python
r_value=dtc.score(x,y)
print('r_square_value :',r_value*100)
```
